```python
import cv2
import matplotlib.pyplot as plt
import scipy.io
```

```python
data = scipy.io.loadmat("annotations/anno_train.mat")
anno_train = data["anno_train_aligned"][0]
```

```python
city_names = [item["cityname"][0][0][0] for item in anno_train]
image_names = [item["im_name"][0][0][0] for item in anno_train]
annotations = [item["bbs"][0][0] for item in anno_train]
```

```python
image_path = "leftImg8bit_trainvaltest/leftImg8bit/"
idx = 5
image = cv2.imread(f"{image_path}train/{city_names[idx]}/{image_names[idx]}")
# Convert the image from BGR to RGB for displaying purposes
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
bboxes = annotations[idx][:, 0:5]
```

```python
for bbox in bboxes:
    class_label, x1, y1, w, h = bbox
    # Draw the bounding box rectangle on the image
    cv2.rectangle(image, (x1, y1), (x1 + w, y1 + h), (0, 255, 0), 2)
    # Display the class label
    cv2.putText(image, str(class_label), (x1, y1 - 5), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
```

```python
plt.imshow(image)
plt.show()
```
