```python
import re 
import os

def open_file(filename):
    with open(filename, 'r') as f:
        content = f.read()
        f.close()
    return content

def clean_content(content):
    content = re.split(r'\n\n', content)
    content = list(filter(lambda x : 'ms inference' in x, content))
    content = list(map(lambda x : re.split(r'\n', x), content))
    #print(content[0])
    content = list(map(lambda x : x[0] + " inference, " + x[1], content))
    #print(content[0])
    content_col_name = list(map(lambda x : re.split(r"\n|,\s|', '|per image at shape ", x), content))
    #print(content_col_name[0])
    #print(content_col_name[0])
    content_col_name = list(map(lambda x : list(map(lambda y : re.sub(r'\d+|\.|Speed: ','', y), x)), content_col_name))
    #print(content_col_name[0])
    content_col_name = list(map(lambda x : list(map(lambda y : re.split(r'\s', y), x)), content_col_name))
    #print(content_col_name[0])
    content_col_name = list(map(lambda x : list(map(lambda y : list(filter(lambda z : len(z) > 1 and z != 'x' and z != '' and z != 'ms', y)), x)), content_col_name))
    #print(content_col_name[0])
    content_col_name = list(map(lambda x : list(map(lambda y : " ".join(y), x)), content_col_name))
    #print(content_col_name[0])
    def f(x):
        x[-1] = "width"
        x[-2] = "height"
        x[-3] = "channels"
        x[-4] = "batch size"
        x[-8] = "inference"
        return x
    try:
        content_col_name = list(map(lambda x : f(x), content_col_name))
    except:
        for i,x in enumerate(content_col_name):
            try:
                f(x)
            except:
                print(i, x)
    content_data = list(map(lambda x : re.split(r"\n|,\s|640x640\s|ms postprocess per image at shape |', '", x), content))
    #print(content_data[0])
    content_data = list(map(lambda x : list(map(lambda y : re.sub(r'(\s|:|\^|[A-Za-z:()])+', '', y), x)), content_data))
    content_data = list(map(lambda x : list(filter(lambda y : y != '', x)), content_data))
    #print(content_data[0])
    content_data = list(map(lambda x : list(map(lambda y : float(y) if "." in y else int(y), x[1:])), content_data))
    out = list(map(lambda x, y : list(map(lambda a,b : (a,b), x,y)), content_col_name, content_data))
    #for i,x in out:
     #   assert 
    return out
#def write_file(filename, content):



#print(open_file('jetson.txt'))
X = clean_content(open_file('jetson_ncnn_short.txt'))
#filenames = ['jetson.txt',  'pi1.txt',  'pi2.txt',  'piOs.txt',  'piOs2.txt', 'piUbuntu.txt', 'raspberryLite.txt']
filenames = os.listdir('.')
filenames = list(filter(lambda x : x.endswith('.txt'), filenames))
print(filenames)
#open_file(f"../data/{filenames[-1]}")
data = {}
for x in filenames:
    try:
        data[x] = clean_content(open_file(f"{x}"))
    except Exception as e:
        print(f"{x} {e}")
```

```python
def write_file(filename, content, type = 'csv'):
    with open(filename, 'w') as f:
        if type == 'csv':
            import csv
            writer = csv.writer(f)
            writer.writerows(content)
        elif type == 'json':
            import json
            f.write(json.dumps(content))
        f.close()

#write_file('jetson.csv', X, 'csv')
```

```python
import pandas as pd
c.execute("DROP TABLE IF EXISTS measurements_detections")
c.execute("""
CREATE TABLE IF NOT EXISTS measurements_detections AS
SELECT device, observation, model, video,
MAX((CASE WHEN measurement = 'pedestrians' THEN CAST(value as INTEGER) ELSE NULL END)) AS pedestrians,
MAX((CASE WHEN measurement = 'cars' THEN CAST(value as INTEGER) ELSE NULL END)) AS cars,
MAX((CASE WHEN measurement = 'truck' THEN CAST(value as INTEGER) ELSE NULL  END)) AS truck,
MAX((CASE WHEN measurement = 'traffic lights' THEN CAST(value as INTEGER) ELSE NULL END)) AS traffic_lights,
MAX((CASE WHEN measurement = 'traffic sign' THEN CAST(value as INTEGER) ELSE NULL END)) AS traffic_sign
FROM measurements
GROUP BY device, observation, model, video;""")
c.execute("SELECT * FROM measurements_detections LIMIT 10")
for x in c.fetchall():
    print(x)
X = pd.read_sql_query("SELECT * FROM measurements_detections", conn)
X.to_csv('measurements_detections.csv')
```

```python
c.execute("DROP TABLE IF EXISTS measurements_speed")
c.execute("""
CREATE TABLE IF NOT EXISTS measurements_speed AS 
SELECT device, observation, model, video, 
MAX((CASE WHEN measurement = 'preprocess' THEN value ELSE 0 END)) AS preprocess,
MAX((CASE WHEN measurement = 'inference' THEN value ELSE 0 END)) AS inference,
MAX((CASE WHEN measurement = 'postprocess' THEN value ELSE 0 END)) AS postprocess
FROM measurements
GROUP BY device, observation, model, video;""")
c.execute("SELECT * FROM measurements_speed LIMIT 10")
#for x in c.fetchall():
#    print(x)
c.execute("DROP TABLE IF EXISTS measurements_speed_avg")
c.execute("""
CREATE TABLE IF NOT EXISTS measurements_speed_avg AS     
SELECT 'all-devices' AS device, 'all-models' AS model, 'all-videos' AS video, AVG(m0.preprocess) preprocess_mean, AVG((m0.preprocess - m1.preprocess)*(m0.preprocess - m1.preprocess)) AS preprocess_sd,
MIN(m0.preprocess) AS preprocess_min, MAX(m0.preprocess) AS preprocess_max,
AVG(m0.inference) AS inference_mean, AVG((m0.inference - m1.inference)*(m0.inference - m1.inference)) AS inference_sd,
MIN(m0.inference) AS inference_min, MAX(m0.inference) AS inference_max,
AVG(m0.postprocess) AS postprocess_mean, AVG((m0.postprocess - m1.postprocess)*(m0.postprocess - m1.postprocess)) AS postprocess_sd,
MIN(m0.postprocess) AS postprocess_min, MAX(m0.postprocess) AS postprocess_max
FROM measurements_speed m0, (
SELECT AVG(preprocess) AS preprocess, AVG(inference) AS inference, AVG(postprocess) AS postprocess
FROM measurements_speed) m1
UNION
SELECT 'all-devices' AS device, model, 'all-videos' AS video, AVG(m0.preprocess) preprocess_mean, AVG((m0.preprocess - m1.preprocess)*(m0.preprocess - m1.preprocess)) AS preprocess_sd,
MIN(m0.preprocess) AS preprocess_min, MAX(m0.preprocess) AS preprocess_max,
AVG(m0.inference) AS inference_mean, AVG((m0.inference - m1.inference)*(m0.inference - m1.inference)) AS inference_sd,
MIN(m0.inference) AS inference_min, MAX(m0.inference) AS inference_max,
AVG(m0.postprocess) AS postprocess_mean, AVG((m0.postprocess - m1.postprocess)*(m0.postprocess - m1.postprocess)) AS postprocess_sd,
MIN(m0.postprocess) AS postprocess_min, MAX(m0.postprocess) AS postprocess_max
FROM measurements_speed m0, (
SELECT AVG(preprocess) AS preprocess, AVG(inference) AS inference, AVG(postprocess) AS postprocess
FROM measurements_speed) m1
GROUP BY model
UNION
SELECT device, 'all-models' AS model, 'all-videos' AS video, AVG(m0.preprocess) preprocess_mean, AVG((m0.preprocess - m1.preprocess)*(m0.preprocess - m1.preprocess)) AS preprocess_sd,
MIN(m0.preprocess) AS preprocess_min, MAX(m0.preprocess) AS preprocess_max,
AVG(m0.inference) AS inference_mean, AVG((m0.inference - m1.inference)*(m0.inference - m1.inference)) AS inference_sd,
MIN(m0.inference) AS inference_min, MAX(m0.inference) AS inference_max,
AVG(m0.postprocess) AS postprocess_mean, AVG((m0.postprocess - m1.postprocess)*(m0.postprocess - m1.postprocess)) AS postprocess_sd,
MIN(m0.postprocess) AS postprocess_min, MAX(m0.postprocess) AS postprocess_max
FROM measurements_speed m0, (
SELECT AVG(preprocess) AS preprocess, AVG(inference) AS inference, AVG(postprocess) AS postprocess
FROM measurements_speed) m1
GROUP BY device
UNION
SELECT device, model, 'all-videos' AS video, AVG(m0.preprocess) preprocess_mean, AVG((m0.preprocess - m1.preprocess)*(m0.preprocess - m1.preprocess)) AS preprocess_sd,
MIN(m0.preprocess) AS preprocess_min, MAX(m0.preprocess) AS preprocess_max,
AVG(m0.inference) AS inference_mean, AVG((m0.inference - m1.inference)*(m0.inference - m1.inference)) AS inference_sd,
MIN(m0.inference) AS inference_min, MAX(m0.inference) AS inference_max,
AVG(m0.postprocess) AS postprocess_mean, AVG((m0.postprocess - m1.postprocess)*(m0.postprocess - m1.postprocess)) AS postprocess_sd,
MIN(m0.postprocess) AS postprocess_min, MAX(m0.postprocess) AS postprocess_max
FROM measurements_speed m0, (
SELECT AVG(preprocess) AS preprocess, AVG(inference) AS inference, AVG(postprocess) AS postprocess
FROM measurements_speed) m1
GROUP BY device, model
UNION
SELECT device, model, video, AVG(m0.preprocess) preprocess_mean, AVG((m0.preprocess - m1.preprocess)*(m0.preprocess - m1.preprocess)) AS preprocess_sd,
MIN(m0.preprocess) AS preprocess_min, MAX(m0.preprocess) AS preprocess_max,
AVG(m0.inference) AS inference_mean, AVG((m0.inference - m1.inference)*(m0.inference - m1.inference)) AS inference_sd,
MIN(m0.inference) AS inference_min, MAX(m0.inference) AS inference_max,
AVG(m0.postprocess) AS postprocess_mean, AVG((m0.postprocess - m1.postprocess)*(m0.postprocess - m1.postprocess)) AS postprocess_sd,
MIN(m0.postprocess) AS postprocess_min, MAX(m0.postprocess) AS postprocess_max
FROM measurements_speed m0, (
SELECT AVG(preprocess) AS preprocess, AVG(inference) AS inference, AVG(postprocess) AS postprocess
FROM measurements_speed) m1
GROUP BY device, model, video
;""")
c.execute("SELECT * FROM measurements_speed_avg LIMIT 10")
#for x in c.fetchall():
#    print(x)
c.execute("DROP VIEW IF EXISTS measurements_speed_avg_view")
c.execute("CREATE VIEW IF NOT EXISTS measurements_speed_avg_view AS SELECT device, model, video, round(preprocess_mean, 2) as preprocess_mean, round(preprocess_sd, 2) as preprocess_sd, preprocess_min, preprocess_max, round(inference_mean, 2) as inference_mean, round(inference_sd, 2) as inference_sd, inference_min, inference_max, round(postprocess_mean, 2) as postprocess_mean, round(postprocess_sd, 2) as postprocess_sd, postprocess_min, postprocess_max FROM measurements_speed_avg")
c.execute("SELECT * FROM measurements_speed_avg_view LIMIT 50")
for x in c.fetchall():
    print(x)
X = pd.read_sql_query("SELECT * FROM measurements_speed_avg_view", conn)
X.to_csv('measurements_speed_avg.csv')
```

```python
c.execute("DROP TABLE IF EXISTS measurements_shape")
c.execute("""
CREATE TABLE IF NOT EXISTS measurements_shape AS SELECT device, model, video, COUNT(*) as count, 
width,  SUM(width) = COUNT(*) * MAX(width) as all_outputs_same_width, 
height, SUM(height) = COUNT(*) * MAX(height) as all_outputs_same_height,
channels, SUM(channels) = COUNT(*) * MAX(channels) as all_outputs_same_channels,
batch_size, SUM(batch_size) = COUNT(*) * MAX(batch_size) as all_outputs_same_batch_size
FROM (SELECT device, model, video, observation,
MAX((CASE WHEN measurement = 'width' THEN value ELSE 0 END)) AS width,
MAX((CASE WHEN measurement = 'height' THEN value ELSE 0 END)) AS height,
MAX((CASE WHEN measurement = 'channels' THEN value ELSE 0 END)) AS channels,
MAX((CASE WHEN measurement = 'batch size' THEN value ELSE 0 END)) AS batch_size
FROM measurements
GROUP BY observation, device, model, video) 
GROUP BY device, model, video, width, height, channels, batch_size
;""")
c.execute("SELECT * FROM measurements_shape")
#for x in c.fetchall():
#    print(x)
X = c.fetchall()
for x in X:
    print(x)
#write_file('measurements_shape.csv', X)
X = pd.read_sql_query("SELECT * FROM measurements_shape", conn)
X.to_csv('measurements_shape.csv')
```
